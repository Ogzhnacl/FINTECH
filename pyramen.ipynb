```python

#Read in the menu and sales data menu_data.csv and sales_data.csv
import pandas as pd 
import numpy as np
menu_reader = pd.read_csv('menu_data.csv')
sales_reader = pd.read_csv('sales_data.csv')
```

```python
#check the data 
menu_reader.head()
```

```python
#Rename the Item column from item to Menu_Item to merge menu_data.csv and sales_data.csv
menu_reader=menu_reader.rename({'item':'Menu_Item'}, axis='columns')
```

```python
#check the data
menu_reader.head()
```

```python
#Merge the data on Menu Item
dfjoin = pd.merge(sales_reader,menu_reader, on='Menu_Item', how='inner')
dfjoin.head()
```

```python
#Calculate and add columns COGS , Profit , revenue
dfjoin['COGS'] = (dfjoin['Quantity'] * dfjoin['cost'])
dfjoin['Profit'] = (dfjoin['price'] - dfjoin['cost']) * dfjoin['Quantity']
dfjoin['revenue'] = (dfjoin['Quantity'] * dfjoin['price'])
dfjoin.head()
```

```python
#Group and sum COGS , Profit , revenue values according to the menu item
dfjoin=dfjoin.groupby(by=['Menu_Item']).sum().reset_index()
dfjoin = dfjoin.drop(columns=['Line_Item_ID','Credit_Card_Number','price'])
```

```python
dfjoin
```

```python
#define report dict to hold our key-value pairs of items and metrics
report = {}
for i in dfjoin['Menu_Item']:
    report[i] = [{"01-count":dfjoin['Quantity'][j],'02-revenue':dfjoin['revenue'][j],'03-cogs':dfjoin['COGS'][j],'04-profit':dfjoin['Profit'][j]} for j in dfjoin[dfjoin['Menu_Item']==i].index]
```

```python
#Write out report to a text file
with open('Report.txt',"w") as file:
    file.write(f"Report:")
    file.write("\n")
    for line in report:
        file.write(f"{line}:{report[line]}")
        file.write("\n")
```
